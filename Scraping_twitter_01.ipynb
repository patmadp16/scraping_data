```python
#from tweepy.streaming import StreamListener
import tweepy
from time import sleep
from tweepy import OAuthHandler
from tweepy import Stream
import datetime
import json, time, sys
import csv


# Inisialisasi API Key yang terdaftar

consumer_key = ""
consumer_secret = ""
access_token = ""
access_token_secret = ""

# Autentikasi API Twitter menggunakan Tweepy
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth,wait_on_rate_limit=True)

```

```python
from typing import Counter

header = ['tanggal','Text']
with open('/content/drive/MyDrive/NLP/pejabat_negara.csv', 'w', encoding='UTF8', newline='') as f:
    writer = csv.writer(f)
    i=0
    # write the header
    writer.writerow(header)
    max_tweets = 500
    # cari tweet yang berbahasa indonesia
    for tweet in tweepy.Cursor(api.user_timeline,screen_name="@jokowi", tweet_mode='extended').items(500):
        #menampilkan hasil 
        print(i, "\n NGETWEET \n",tweet.created_at, tweet.full_text,"\n")
        i=i+1
        #mnyimpan data dalam row
        writer.writerow([tweet.created_at, tweet.full_text])
```
